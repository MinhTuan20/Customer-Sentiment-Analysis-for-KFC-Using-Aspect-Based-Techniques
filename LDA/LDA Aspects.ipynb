## <div style="border-radius:10px; border:#D07220 solid; padding: 12px; background-color: #Q07220; font-size:100%; font-family:Comic Sans MS; font-family:Comic Sans MS; text-align:center"> REMOVE STOPWORD </div>

```python
import nltk
nltk.download('punkt')  # Để sử dụng tokenizer
nltk.download('stopwords')  # Tải stopwords
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
from nltk.tokenize import word_tokenize

def remove_stopwords(comment):
    if isinstance(comment, str):
        word_tokens = word_tokenize(comment.lower())
        filtered_words = [word for word in word_tokens if word not in stop_words and word.isalnum()]
        return " ".join(filtered_words)
    else:
        return ""  
```

```python
print(google_df['translated_text'].dtype)
google_df['translated_text'] = google_df['translated_text'].astype(str).fillna('')
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

google_df['cleaned_translated_text'] = google_df['translated_text'].apply(remove_stopwords)

print(google_df[['translated_text', 'cleaned_translated_text']].head())
```

## <div style="border-radius:10px; border:#D07220 solid; padding: 12px; background-color: #Q07220; font-size:100%; font-family:Comic Sans MS; font-family:Comic Sans MS; text-align:center"> REMOVE EMOJIS, SPEACIAL AND UNWANTED CHARACTERS </div>

```python
import re

# Hàm để xóa emoji, ký tự đặc biệt, và các ký tự không mong muốn
def remove_special_characters(comment):
    # Xóa emoji
    comment = re.sub(r'[^\x00-\x7F]+', '', comment)
    # Xóa các ký tự đặc biệt, dấu gạch đầu dòng, dấu cách thừa
    comment = re.sub(r'[^a-zA-Z0-9\s]', '', comment)
    # Xóa các khoảng trắng dư thừa
    comment = re.sub(r'\s+', ' ', comment).strip()
    return comment

# Áp dụng hàm remove_special_characters vào cột 'cleaned_translated_text'
google_df['final_cleaned_text'] = google_df['cleaned_translated_text'].apply(remove_special_characters)

# Kiểm tra kết quả
print(google_df[['cleaned_translated_text', 'final_cleaned_text']].head())
```

## <div style="border-radius:10px; border:#D07220 solid; padding: 12px; background-color: #Q07220; font-size:100%; font-family:Comic Sans MS; font-family:Comic Sans MS; text-align:center"> FILTER COMMENTS LESS THAN 3 WORDS </div>

```python
# Hàm lọc các bình luận có ít hơn 3 từ
def remove_short_comments(comment, min_words=3):
    # Kiểm tra số lượng từ trong bình luận
    if len(comment.split()) < min_words:
        return None  # Nếu bình luận có ít hơn 3 từ, trả về None (sẽ bị loại bỏ sau)
    return comment

# Áp dụng hàm lọc vào cột 'final_cleaned_text'
google_df['filtered_comments'] = google_df['final_cleaned_text'].apply(lambda x: remove_short_comments(x))

# Loại bỏ các dòng có giá trị None (bình luận ít hơn 3 từ)
google_df = google_df.dropna(subset=['filtered_comments'])

# Kiểm tra kết quả
print(google_df[['final_cleaned_text', 'filtered_comments']].head())
```

```python
google_df = google_df.rename(columns={'filtered_comments': 'processed_text'})

google_df = google_df[['user_name','rating', 'translated_text', 'processed_text']]

google_df
```

## <div style="border-radius:10px; border:#D07220 solid; padding: 12px; background-color: #Q07220; font-size:100%; font-family:Comic Sans MS; font-family:Comic Sans MS; text-align:center"> TOKENIZATION </div>

```python
def tokenize_text(text):
    if isinstance(text, str):
        return word_tokenize(text.lower()) 
    return []

google_df['processed_text_Final'] = google_df['processed_text'].apply(tokenize_text)

print(google_df[['processed_text', 'processed_text_Final']].head())
```

# <div style="border-radius:10px; border:#D07220 solid; padding: 12px; background-color: #Q07220; font-size:100%; font-family:Comic Sans MS; font-family:Comic Sans MS; text-align:center"> LDA </div>

```python
from gensim import corpora

dictionary = corpora.Dictionary(google_df['processed_text_Final'])

corpus = [dictionary.doc2bow(text) for text in google_df['processed_text_Final']]

print(corpus[:2])
```

```python
print(f"Số lượng từ khóa duy nhất: {len(dictionary)}")
print(f"Số lượng tài liệu: {len(corpus)}")
```

```python
from gensim.models import Phrases, LdaModel, CoherenceModel
from gensim.models.phrases import Phraser
from gensim import corpora
import matplotlib.pyplot as plt
import pyLDAvis.gensim_models as gensimvis
import pyLDAvis
from textblob import TextBlob
```

## <div style="border-radius:10px; border:#D07220 solid; padding: 12px; background-color: #Q07220; font-size:100%; font-family:Comic Sans MS; font-family:Comic Sans MS; text-align:center"> COHERENCE SCORE </div>

```python
coherence_scores = []
for num_topics in range(5, 21):  # Thử từ 5 đến 20 chủ đề
    lda_model = LdaModel(corpus=corpus,
                         id2word=dictionary,
                         num_topics=num_topics,
                         random_state=42,
                         passes=20,
                         alpha='asymmetric',
                         eta='auto')
    
    coherence_model = CoherenceModel(model=lda_model, texts=google_df['processed_text_Final'], dictionary=dictionary, coherence='c_v')
    score = coherence_model.get_coherence()
    coherence_scores.append((num_topics, score))

for num, score in coherence_scores:
    print(f"Chủ đề: {num} - Coherence Score: {score:.4f}")
```

## <div style="border-radius:10px; border:#D07220 solid; padding: 12px; background-color: #Q07220; font-size:100%; font-family:Comic Sans MS; font-family:Comic Sans MS; text-align:center"> 7 TOPICS LDA </div>

```python
lda_5_models = LdaModel(corpus=corpus,
                          id2word=dictionary,
                          num_topics= 5,
                          random_state=42,
                          passes=30,
                          alpha='auto',
                          eta=0.005)

pyLDAvis.enable_notebook()
lda_vis_data_2 = gensimvis.prepare(lda_5_models, corpus, dictionary,mds='mmds', lambda_step=0.1)
pyLDAvis.display(lda_vis_data_2)
```

```python
for idx, topic in lda_5_models.print_topics(-1):
    print(f"Topic {idx}: {topic}\n")
```
